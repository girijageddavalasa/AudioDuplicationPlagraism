```python
import pretty_midi
from basic_pitch.inference import predict
import pandas as pd  # Make sure pandas is imported
import os            # Make sure os is imported

SAMPLES = ['/kaggle/input/samsung5thnov/sample1.wav', '/kaggle/input/samsung5thnov/sample2.wav', '/kaggle/input/samsung5thnov/sample3.wav']
sample_dfs = []

print(f"--- Starting to process {len(SAMPLES)} audio files ---")

for s in SAMPLES:
    print(f"\nProcessing file: {s}")
    
    # 1. Run the prediction
    print(f"Running basic_pitch predict() on {os.path.basename(s)}...")
    out = predict(s)
    print("...Prediction complete.")

    # 2. Extract events
    events = out[2]  # Assuming out[2] contains the note events
    print(f"Found {len(events)} note events.")

    # 3. Create and modify the DataFrame
    df = pd.DataFrame(events, columns=['start', 'end', 'pitch_midi', 'velocity', 'mask'])
    df['track'] = os.path.basename(s)
    df['note_name'] = df['pitch_midi'].apply(pretty_midi.note_number_to_name)
    df['genre'] = 'sample'
    
    print(f"Created DataFrame for {os.path.basename(s)}.")
    
    # <--- NEW! Here is the 'table' you asked for (the first 5 rows)
    print(f"--- Attribute table (head) for {os.path.basename(s)}: ---")
    print(df.head())
    print("--------------------------------------------------")
    
    # 4. Append to the list
    sample_dfs.append(df)

print("\n--- All files processed. Ready to concatenate DataFrames. ---")

# You can add this part to see the final combined DataFrame
if sample_dfs:
    final_df = pd.concat(sample_dfs, ignore_index=True)
    print("Final combined DataFrame created.")
    print(final_df.head())
else:
    print("No DataFrames were created.")
```

```python
import os
import pandas as pd
import time

# --- NEW, BETTER SIMILARITY FUNCTION ---
def containment_similarity(sample_notes, db_notes):
    """
    Calculates what percentage of the sample's unique notes
    are *contained within* the database track's unique notes.
    """
    # Create sets of unique notes
    sample_set = set(sample_notes)
    db_set = set(db_notes)
    
    # If basic_pitch found no notes, similarity is 0
    if not sample_set:
        return 0.0, set() 
    
    # Find the notes from the sample that are ALSO in the database
    intersection = sample_set & db_set
    
    # This is the new, more accurate score:
    # (Common Notes) / (Total Unique Notes in Sample)
    score = len(intersection) / len(sample_set)
    
    return score, intersection
# --- END OF NEW FUNCTION ---


REPORTS_PATH = '/kaggle/working/plagiarism_reports'
os.makedirs(REPORTS_PATH, exist_ok=True)
print(f"--- Reports directory ensured at: {REPORTS_PATH} ---")

total_start_time = time.time()

# (Assuming 'sample_dfs' and 'corpus_df' exist from previous cells)
for sample_df in sample_dfs:
    sample_track_name = sample_df['track'].iloc[0]
    
    print(f"\n========================================================")
    print(f"Processing Sample File: {sample_track_name}")
    print(f"========================================================")
    
    sample_seq = sample_df['note_name'].tolist()
    
    print(f"-> Sample sequence has {len(sample_seq)} total notes ({len(set(sample_seq))} unique).")
    print(f"-> Now comparing against all tracks in the corpus...")
    
    rows = []
    sample_start_time = time.time()
    
    for (genre, track), db_df in corpus_df.groupby(['genre', 'track']):
        
        # This print statement updates on one line
        print(f"  ... comparing against: [Genre: {genre}, Track: {track}]", end='\r')
        
        db_seq = db_df['note_name'].tolist()
        
        # --- THE ONLY CHANGE IS HERE ---
        # We now call our new, more accurate function
        sim, common_notes = containment_similarity(sample_seq, db_seq)
        # --- END OF CHANGE ---
        
        flag = sim >= 0.7  # Your 70% threshold

        if flag:
            # Print a newline to not overwrite the progress bar
            print(f"\n--- !!! HIGH SIMILARITY FOUND !!! ---")
            print(f"    Sample:   {sample_track_name}")
            print(f"    DB Track: {track} (Genre: {genre})")
            print(f"    Similarity: {sim*100:.2f}% (containment)")
            print(f"    Common notes ({len(common_notes)}): {sorted(list(common_notes))}")
            print(f"--------------------------------------")
        
        rows.append({
            'sample': sample_track_name,
            'db_track': track,
            'db_genre': genre,
            'similarity_%': f"{sim*100:.2f}",
            'plagiarism_flag': flag
        })
    
    sample_end_time = time.time()
    print() # Print a blank line to clear the "comparing..." line
    print(f"-> Finished all comparisons for {sample_track_name} in {sample_end_time - sample_start_time:.2f} seconds.")
    
    result_df = pd.DataFrame(rows)
    print(f"\n--- Properly Calculated Report for {sample_track_name}: ---")
    # Sort by the new, more meaningful similarity score
    print(result_df.sort_values('similarity_%', ascending=False).head(5))
    
    report_path = os.path.join(REPORTS_PATH, f'plagiarism_report_{sample_track_name}.xlsx')
    result_df.to_excel(report_path, index=False)
    print("Full report saved:", report_path)

total_end_time = time.time()
print(f"\n--- All sample files have been processed. Total runtime: {total_end_time - total_start_time:.2f} seconds. ---")
```
